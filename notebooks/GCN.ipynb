```python
import torch
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.loader import NeighborLoader
from torch_geometric.nn import GCNConv
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

feature_cols = ['cfo2cl', 'dta2ev', 'NNP_SD', 'ocfa4q', 'r_nta', 'roic4q', 'yoy_s']
target_col = 'return'

df['return'] = df['return'].clip(lower=-0.5, upper=0.5)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[feature_cols].values)
y = df[target_col].values

k = 5

knn = NearestNeighbors(n_neighbors=k + 1, n_jobs=-1)
knn.fit(X_scaled)
distances, indices = knn.kneighbors(X_scaled)

source_nodes = np.repeat(np.arange(len(X_scaled)), k)
target_nodes = indices[:, 1:].flatten()
edge_index_np = np.stack([source_nodes, target_nodes], axis=0)
edge_index = torch.tensor(edge_index_np, dtype=torch.long)


data = Data(
    x=torch.tensor(X_scaled, dtype=torch.float32),
    edge_index=edge_index,
    y=torch.tensor(y, dtype=torch.float32)
)

rk_cutoff = df['rk'].quantile(0.8)

train_mask_indices = df[df['rk'] <= rk_cutoff].index
test_mask_indices = df[df['rk'] > rk_cutoff].index

train_loader = NeighborLoader(
    data,
    num_neighbors=[10, 5],
    batch_size=4096,
    input_nodes=torch.tensor(train_mask_indices),
    shuffle=True
)

test_loader = NeighborLoader(
    data,
    num_neighbors=[10, 5],
    batch_size=4096,
    input_nodes=torch.tensor(test_mask_indices),
    shuffle=False
)

class GCN(torch.nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.conv1 = GCNConv(input_dim, 32)
        self.conv2 = GCNConv(32, 1)

    def forward(self, x, edge_index):
        x = F.relu(self.conv1(x, edge_index))
        x = self.conv2(x, edge_index)
        return x.squeeze()

LEARNING_RATE = 0.001
EPOCHS = 10

model = GCN(input_dim=X_scaled.shape[1])
optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)

# --- Training Loop ---
model.train()
print(f"Starting training with LR={LEARNING_RATE}...")

for epoch in range(EPOCHS):
    total_loss = 0
    num_batches = 0
    for batch in train_loader:
        optimizer.zero_grad()
        out = model(batch.x, batch.edge_index)

        batch_curr = batch.batch_size
        out_sliced = out[:batch_curr]
        y_sliced = batch.y[:batch_curr]

        loss = F.mse_loss(out_sliced, y_sliced.float())
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        num_batches += 1

        if epoch == 0 and num_batches == 1:
            print(f"\n--- [DEBUG] Sample Check (Epoch 1) ---")
            print(f"Preds: {out_sliced[:5].detach().cpu().numpy()}")
            print(f"True : {y_sliced[:5].detach().cpu().numpy()}")
            print(f"------------------------------------\n")

    avg_loss = total_loss / num_batches
    print(f"Epoch {epoch+1}, Avg Batch Loss: {avg_loss:.5f}")
```

```python
from scipy.stats import spearmanr

model.eval()
all_preds = []
all_true = []

# 1. Get predictions for the test dataset
with torch.no_grad():
    for batch in test_loader:
        out = model(batch.x, batch.edge_index)
        batch_size = batch.batch_size
        pred = out[:batch_size]
        true = batch.y[:batch_size]

        all_preds.append(pred.cpu().numpy())
        all_true.append(true.cpu().numpy())

# 2. Flatten lists
all_preds = np.concatenate(all_preds)
all_true = np.concatenate(all_true)

# 3. Calculate Spearman Correlation
ic, p_value = spearmanr(all_preds, all_true)

print(f"--- Model Performance ---")
print(f"Rank IC: {ic:.4f}")
print(f"P-Value: {p_value:.4f}")

if ic > 0.03:
    print("Result: Strong predictive power.")
elif ic > 0:
    print("Result: Weak positive predictive power.")
else:
    print("Result: No predictive power (or negative correlation).")
```
