```python
import pandas as pd

df1 = pd.read_feather('cfo2cl.ft')
df2 = pd.read_feather('dta2ev.ft')
df3 = pd.read_feather('NNP_SD.ft')
df4 = pd.read_feather('ocfa4q.ft')
df5 = pd.read_feather('r_nta.ft')
df6 = pd.read_feather('roic4q.ft')
df7 = pd.read_feather('yoy_s.ft')
```

```python
from functools import reduce

dfs = {
    'cfo2cl': df1,
    'dta2ev': df2,
    'NNP_SD': df3,
    'ocfa4q': df4,
    'r_nta': df5,
    'roic4q': df6,
    'yoy_s': df7
}

cleaned_dfs = {}

for name, df in dfs.items():
    df = df[['rk', 'code', 'value']]
    df = df.rename(columns={'value': name})
    cleaned_dfs[name] = df

merged = reduce(
    lambda left, right: left.merge(right, on=['rk', 'code'], how='outer'),
    cleaned_dfs.values()
)

merged.head(), merged.shape
```

```python
#drop na in merged
merged = merged.dropna()
#save the file to a certain path
merged.to_feather('merged_factors.ft')
```
